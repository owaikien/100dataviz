### Data Visualisation

Plotting average price for different brand names

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.ticker import MaxNLocator
```

```python
avg_price = df.groupby('brand_name')['price'].mean().reset_index()
avg_price = avg_price.sort_values(by='price', ascending=False).reset_index(drop=True)

# filter top 10 phones with highest average price
top10_avg_price = avg_price.head(n=10)
```

```python
top10_avg_price
```

#### Bar Plot

```python
# Create the figure and objects, specify size and dots per inches
fig, ax = plt.subplots(figsize=(13.33, 7.5), dpi=96)

# Create the grid 
ax.grid(which="major", axis='x', color='#DAD8D7', alpha=0.5, zorder=1)
ax.grid(which="major", axis='y', color='#DAD8D7', alpha=0.5, zorder=1)

# Reformat x-axis label and tick labels
ax.set_xlabel('Smartphone Brands', fontsize=12, labelpad=7) # No need for an axis label
ax.xaxis.set_label_position("bottom")
ax.xaxis.set_tick_params(pad=2, labelbottom=True, bottom=True, labelsize=12, labelrotation=0)

# Set custom labels for x-axis ticks
labels = [e.title() for e in top10_avg_price['brand_name']]
ax.set_xticks(range(len(labels)))  # Set tick positions
ax.set_xticklabels(labels)  # Set tick labels

# Reformat y-axis
ax.set_ylabel('Price ($)', fontsize=12, labelpad=10)
ax.yaxis.set_label_position("left")
ax.yaxis.set_major_formatter(lambda s, i : f'{s:,.0f}')
ax.yaxis.set_major_locator(MaxNLocator(integer=True))
ax.yaxis.set_tick_params(pad=2, labeltop=False, labelbottom=True, bottom=False, labelsize=12)

# Colours - Choose the extreme colours of the colour map
colours = ["#2196f3", "#bbdefb"]

# Colormap - Build the colour maps
cmap = mpl.colors.LinearSegmentedColormap.from_list("colour_map", colours, N=256)
norm = mpl.colors.Normalize(top10_avg_price['price'].min(), top10_avg_price['price'].max()) # linearly normalizes data into the [0.0, 1.0] interval

# Add labels on top of each bar
bar1 = ax.bar(top10_avg_price['brand_name'], top10_avg_price['price'], color=cmap(norm(top10_avg_price['price'])), width=0.6, zorder=2)
ax.bar_label(bar1, labels=[f'{e:,.1f}' for e in top10_avg_price['price']], padding=3, color='black', fontsize=8)

# Remove the spines
ax.spines[['top','left','bottom']].set_visible(False)

# Make the left spine thicker
ax.spines['right'].set_linewidth(1.1)

# Add in red line and rectangle on top
ax.plot([0.12, .9], [.98, .98], transform=fig.transFigure, clip_on=False, color='#E3120B', linewidth=.6)
ax.add_patch(plt.Rectangle((0.12,.98), 0.04, -0.02, facecolor='#E3120B', transform=fig.transFigure, clip_on=False, linewidth = 0))

# Add in title and subtitle
ax.text(x=0.12, y=.93, s="Top 10 Smartphone Brands with the Highest Average Price", transform=fig.transFigure, ha='left', fontsize=14, weight='bold', alpha=.8)
#ax.text(x=0.12, y=.90, s="Exploring the Elites Players in the Smartphone Market", transform=fig.transFigure, ha='left', fontsize=12, alpha=.8)

# Set source text
ax.text(x=0.1, y=0.12, s="Source: Kaggle - Real World Smartphone's Dataset - https://www.kaggle.com/datasets/abhijitdahatonde/real-world-smartphones-dataset", transform=fig.transFigure, ha='left', fontsize=10, alpha=.7)

# Adjust the margins aronund the plot 
plt.subplots_adjust(left=None, bottom=0.2, right=None, top=0.85, wspace=None, hspace=None)

# Set a white background
fig.patch.set_facecolor('white')
```

#### Scatter Plot

Find out if phones with higher price have higher ratings

```python
plt.scatter(df['price'], df['avg_rating'], marker='o', color='black', s=18, edgecolors='black')
plt.xlabel('Price')
plt.ylabel('Average Rating')
plt.title("Relationship between Smartphone Prices and its Ratings")

# Add Grids
plt.grid(True, alpha=0.3)

# Label Vertu smartphone
highlight_index = df[(df['price'] > 5000) & (df['avg_rating'] < 8)].index[0]
highlight_price = df.loc[highlight_index, 'price']
highlight_rating = df.loc[highlight_index, 'avg_rating']
plt.annotate(f'High Price, \nLow Rating', xy=(highlight_price, highlight_rating),
             xytext=(highlight_price + 750, highlight_rating - 0.3),
             arrowprops=dict(facecolor='red', arrowstyle='->'),
             fontsize=10, color='black')

plt.show()
```

#### Correlation Coefficient

```python
# Drop categorical columns: avg_rating, brand_name, model, processor_brand, os
# Drop columns with many missing values: fast_charging
filtered_df = df.drop(columns=['avg_rating', 'brand_name', 'model', 'processor_brand', 'os', 'fast_charging'])

# Remove missing values for the remaining columns
filtered_df.dropna(inplace=True)

# Check
filtered_df.info()
```

```python
# Create correlation heatmap with price
import seaborn as sns
plt.figure(figsize=(8, 12))

heatmap = sns.heatmap(filtered_df.corr()[['price']].sort_values(by='price', ascending=False), vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Features Correlating with Smartphone Price', fontdict={'fontsize': 18}, pad=16)
```
