### Data Visualisation

Plotting average price for different brand names

```python
import matplotlib.pyplot as plt
```

```python
avg_price = df.groupby('brand_name')['price'].mean().reset_index()
avg_price = avg_price.sort_values(by='price', ascending=False).reset_index(drop=True)
```

```python
avg_price['price'] = avg_price['price'] / 100

plt.bar(avg_price['brand_name'], avg_price['price'])
plt.xlabel('Brand Names')
plt.ylabel('Price in USD ($)')
plt.title("Average Price ($) of Different Smartphone Brands")
```

Plotting the average ratings for different brand names

```python
avg_rating = df.groupby('brand_name')['avg_rating'].mean().reset_index()
avg_rating = avg_rating.sort_values(by='avg_rating', ascending=False).reset_index(drop=True)
```

```python
# Plot
top_10_avg = avg_rating.head(n=10)


plt.bar(avg_rating['brand_name'], avg_rating['avg_rating'])
plt.xlabel("Brand Name")
plt.ylabel("Average Rating")
plt.title("Average Rating of Different Smartphone Brands")
plt.show()
```

Find out if phones with higher price have higher ratings

```python
plt.scatter(df['price']/100, df['avg_rating'])
plt.xlabel('Price')
plt.ylabel('Average Rating')
plt.title("Relationship between Smartphone Prices and its Ratings")
```
